```python
import numpy as np
import matplotlib.pyplot as plt

def gradient_descent(X, y, learning_rate, epochs):
    m = 0
    b = 0
    n = len(X)
    costs = []
    
    for epoch in range(epochs):
        y_pred = m * X + b
        error = y_pred - y
        m -= learning_rate * (1/n) * np.sum(X * error)
        b -= learning_rate * (1/n) * np.sum(error)
        
        if epoch % 100 == 0:
            cost = (1/(2*n)) * np.sum(error**2)
            costs.append(cost)
            print(f'Epoch {epoch}, Cost: {cost}')
    
    # Plot epoch vs. cost
    plt.plot(range(0, epochs, 100), costs, marker='o')
    plt.xlabel('Epoch')
    plt.ylabel('Cost')
    plt.title('Epoch vs. Cost')
    plt.show()
    
    return m, b

# Generate some random data for demonstration
np.random.seed(42)
X = np.random.rand(50)
y = 2 + 3 * X + np.random.randn(50) * 0.05

learning_rate = 0.1
epochs = 1000

m, b = gradient_descent(X, y, learning_rate, epochs)
print("Slope (m):", m)
print("Intercept (b):", b)


plt.scatter(X, y, color='blue')  # Plot the data points
plt.plot(X, m * X + b, color='red')  # Plot the best-fit line
plt.xlabel('X')
plt.ylabel('y')
plt.title('Linear Regression with Gradient Descent')
plt.show()
```
